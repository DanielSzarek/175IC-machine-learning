```
import numpy as np
import pandas as pd
import plotly.express as px
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import r2_score
```

```
# Lab
data, _ = make_blobs(n_samples=1000, centers=3, random_state=50)
df = pd.DataFrame(data, columns=['x1', 'x2'])
px.scatter(df, 'x1', 'x2')

cluster = DBSCAN(eps=0.3, min_samples=5)
cluster.fit(data)

df['cluster'] = cluster.labels_
px.scatter(df, 'x1', 'x2', 'cluster', color_continuous_midpoint=0)

df['cluster'] = cluster.labels_
px.scatter(df, 'x1', 'x2', 'cluster', template='plotly_dark', color_continuous_midpoint=0)
```

```
# TASK. Data for clusters
RS = 15
C_STD = 1.1
data, _ = make_blobs(n_samples=1000, centers=4, cluster_std=C_STD, random_state=RS)

df = pd.DataFrame(data, columns=['x1', 'x2'])
px.scatter(df, 'x1', 'x2')
```

```
# DBSCAN
cluster = DBSCAN(eps=0.5, min_samples=4)
cluster.fit(data)

df['cluster'] = cluster.labels_
px.scatter(df, 'x1', 'x2', 'cluster', template='plotly_dark', color_continuous_midpoint=0)
```

```
# K-means
X = df[['x1', 'x2']]
k_means = KMeans(n_clusters=4, random_state=RS).fit(X)

df['k_means'] = k_means.labels_
px.scatter(df, 'x1', 'x2', 'k_means', template='plotly_dark')
```

```
def clusters_comparison(df):
  dbscan_values = df['cluster'].value_counts()
  k_means_values = df['k_means'].value_counts()
  print(f"DBSCAN:\n(CLUSTER AMOUNT):\n {dbscan_values.sort_index()}")
  print(f"KMEANS:\n(CLUSTER AMOUNT):\n {k_means_values.sort_index()}")

clusters_comparison(df)

"""As, you can see, DBSCAN made some mistakes. Clusters 2 and 3 were recognised well.
Unfortunatelly cluster 0 took almost every point which should be added to cluster 1.
There was also many unrecognized points (Cluster -1) and 10 points was added to 5 and 6 clusters.

Kmeans, using Euklides distance made it better. Algorithm arranged every points to 4 clusters and it looks quite well."""
```
