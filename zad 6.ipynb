```
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt

# Task 1. Data preparetion
raw_data = load_iris()
all_data = raw_data.copy()

input_data = all_data['data'][:, :2]
target = all_data['target']

print(input_data[:5])
print(target)
```

```
# Task 2. Data training
X_train, X_test, y_train, y_test = train_test_split(input_data, target)
```

```
# Task 3. SVC

def svc_plot(C=1.0, kernel='rbf'):  # 1.0 and rbf are default values
  svc = SVC(C, kernel)
  svc.fit(X_train, y_train)
  plot_decision_regions(X_train, y_train, clf=svc, legend=2)
  prediction = svc.predict(X_test)
  accuracy = svc.score(X_test, y_test)
  
  plt.xlabel('sepal length (cm)')
  plt.ylabel('sepal width (cm)')
  plt.title(f"SVC accuracy: {accuracy}")
  plt.show()

svc_plot()
```

```
# Task 4. Other SVC kernels
methods = ((0.05, 'rbf'), (0.5, 'linear'), (0.95, 'poly'), (3.0, 'sigmoid'))
for method in methods:
  C, kernel = method
  
  print(f"Method: {kernel}")
  svc_plot(C, kernel)
```
