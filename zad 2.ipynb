```
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

shop_data = {
    'size': ['L', np.nan, 'S', 'XL', 'L', 'M', np.nan, 'S', 'M', 'XL', np.nan, 'M', 'L', np.nan, 'XL', 'S', 'M', 'XL', np.nan, 'L'],
    'color': ['green', np.nan, 'black', 'white', 'blue', np.nan, np.nan, 'black', 'green', 'brown', 'red', 'black', 'white', 'black', 'green', 'red', 'white', 'blue', np.nan, 'pink'],
    'gender': ['female', 'female', 'female', 'male', 'male', 'female', 'male', 'female', np.nan, 'female', 'female', 'male', 'female', 'female', 'male', 'male', 'female', 'female', 'male', 'female'],
    'price': [149.0, 79.0, 70.0, 150.0, 85.0, 92.0, 15.0, 35.50, 49.99, np.nan, 250.0, 153.10, 125.85, 99.99, 100.99, 25.0, 85.0, 65.0, np.nan, 169.0],
    'weight': [380, np.nan, 350, np.nan, 250, 280, 350, 150, 160, np.nan, 210, 250, 300, 750, 500, 450, 175, 100, 125, 350],
    'bought': ['no', 'yes', 'yes', 'no', 'yes', 'yes', 'no', 'yes', 'no', 'no', 'yes', 'no', 'no', 'no', 'yes', 'yes', 'yes', 'no', 'no', 'no']
}


df_shop_data_raw = pd.DataFrame(data=shop_data)
df = df_shop_data_raw.copy()

# Task 1 - remove record if 3 or more values are nan
df.dropna(thresh=len(df.keys())-2, inplace=True)
print(df)

# Task 2-4 - fill empty values in various ways
df['price'].fillna(df['price'].mean(), inplace=True)
df['weight'].fillna(df['weight'].interpolate(), inplace=True)
df['size'].fillna('allsize', inplace=True)
df['color'].fillna('white', inplace=True)
df['gender'].fillna('unisex', inplace=True)
print(df)

# Task 5 - new column with 3 specific class
df['cost'] = pd.cut(df['price'], bins = 3, labels=('cheap', 'normal_price', 'expensive'))
print(df)

# Task 6 - bought_yes column
df['bought_yes'] = df['bought'].apply(lambda x: 1 if 'yes' in x else 0)
print(df)

# Task 7 - get input data and output data then create train and test data
df_input = df[['size', 'color', 'gender', 'cost']]
df_output = df['bought_yes']

x_train, x_test, y_train, y_test = train_test_split(df_input, df_output)
print(x_train)
print(y_train)
```
